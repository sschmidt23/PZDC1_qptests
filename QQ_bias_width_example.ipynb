```python
import numpy as np
import qp
import scipy.stats as ss
import matplotlib.pyplot as plt
import individual_metrics as inmet
%matplotlib inline
```

```python
truefunc = ss.norm(loc=3.0,scale=0.5)
x = np.arange(1.0,8.0,0.02)
truex = truefunc.pdf(x)
fat= ss.norm(loc=3.0,scale=1.0)
fatx = fat.pdf(x)
skinny = ss.norm(loc=3.0,scale=0.25)
skinnyx = skinny.pdf(x)
high = ss.norm(loc=4.0,scale=0.5)
highx = high.pdf(x)
low = ss.norm(loc=2.0,scale=0.5)
lowx = low.pdf(x)
```

```python
fig = plt.figure(figsize=(15,10))
plt.plot(x,truex,lw=4,c='k',label = 'true distribution')
plt.plot(x,fatx,lw=3,c='g',label = 'fat')
plt.plot(x,skinnyx,lw=3,c='r',label='skinny')
plt.plot(x,highx,lw=3,c='m',label='high')
plt.plot(x,lowx,lw=3,c='b',label='low')
plt.legend()
plt.xlabel('redshift')
plt.ylabel('Number')
```

Draw samples of 20,000 objects from each of the distributions and create qp ensemble objects for each of those

```python
truesample = np.array(truefunc.rvs(size=10000))
fatsample = np.array(fat.rvs(size=10000))
skinnysample = np.array(skinny.rvs(size=10000))
highsample = np.array(high.rvs(size=10000))
lowsample = np.array(low.rvs(size=10000))
```

```python
truegrid=np.zeros([20000,len(truex)])
for i in range(20000):
    truegrid[i,:] = truex
```

```python
true_pdf = qp.Ensemble(truegrid.shape[0],gridded=(x,truegrid))
```

```python
testtrue = inmet.EvaluateMetric(true_pdf,truesample)
```

```python
truePIT = testtrue.PIT()
```

```python
testtrue.QQplot(using='gridded',dx=0.001,Nquants=101)
```
